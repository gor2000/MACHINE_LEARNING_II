**C and gamma cherkassky estimations can be computed like this:**

```python
# Load Libraries
%matplotlib inline
import numpy as np
import pandas as pd

# Load data
from sklearn.datasets import load_iris
X = load_iris().data
y = load_iris().target

# Compute dataset statistics
d = X.shape[1]
m = np.mean(y)
s = np.std(y)


# Obtain Cherksassky estimated values
C_cherk = np.max([np.abs(m + 3*s),np.abs(m - 3*s)])
gamma_cherk = np.power(0.35, 1/d)

print("C = " + str(C_cherk) + "- gamma = " + str(gamma_cherk))
```
